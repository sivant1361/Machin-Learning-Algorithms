```python
import pandas as pd
import numpy as n
import matplotlib.pyplot as m
import seaborn as s
import sklearn.metrics as m
from sklearn.linear_model import LinearRegression as l
from sklearn.datasets import load_wine as w
from sklearn.model_selection import train_test_split as t
%matplotlib inline
```

```python
wine=w()
```

```python
wi=pd.DataFrame(wine["data"],columns=wine["feature_names"])
```

```python
wi.head(5)
```

```python
wi.columns
```

```python
x=wi["alcohol"].values.reshape(-1,1)
y=wi["ash"].values.reshape(-1,1)
```

```python
print(x)
```

```python
print(y)
```

```python
xtrain,xtest,ytrain,ytest=t(x,y,random_state=0)
```

```python
reg=l()
reg.fit(xtrain,ytrain)
```

```python
ypred=reg.predict(xtest)
ypred
```

```python
pdf=pd.DataFrame({"Actual":ytest.flatten(),"Predicted":ypred.flatten()})
```

```python
gdf=pdf.head(13).sort_values("Predicted")
gdf
```

```python
s.jointplot(x="Actual",y="Predicted",data=gdf,kind="reg")
```
