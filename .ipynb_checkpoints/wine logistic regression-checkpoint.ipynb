```python
import math as m
import pandas as pd
import numpy as n
from sklearn.linear_model import LogisticRegression as l
from sklearn.model_selection import train_test_split as t
from sklearn.metrics import classification_report as r
from sklearn.metrics import confusion_matrix as m
```

```python
w=pd.read_csv("winequality.csv")
```

```python
w.head()
```

```python
def func(a):
    #if the number is 5.75,0.75 will be stored in k and 5 will be stored in w
    k,w=n.modf(a)
    if k<0.5:
        return w
    elif k>0.5:
        return w+1
```

```python
w["alcohol"]=w["alcohol"].apply(func)
w["alcohol"].head(10)
```

```python
w["fixed acidity"]=w["fixed acidity"].apply(func)
#fill all the NaN with zero
w=w.fillna(0)
```

```python
x=w["density"].values.reshape(-1,1)
```

```python
y=w["fixed acidity"].values.reshape(-1,1)
```

```python
xtr,xte,ytr,yte=t(x,y,test_size=0.2,random_state=0)
```

```python
reg=l()
reg.fit(xtr,ytr)
```

```python
ypred=reg.predict(xte)
```

```python
df=pd.DataFrame({"Actual":yte.flatten(),"Predicted":ypred.flatten()})
```
